```r

	## use library "e1071"
	## use svm() function


	#################################
	## Support Vector Classifier
	####################################

	## kernel="linear" fits support vector classifier
	## cost specifies the cost C to control margin violation

	## simulate a data to fit support vector classifier

	set.seed (1)
	x=matrix (rnorm (20*2) , ncol =2)
	y=c(rep (-1,10) , rep (1 ,10) )
	x[y==1 ,]= x[y==1,] + 1	

	## visualize the data
	plot(x, col =(3-y))

	dat=data.frame(x=x, y=as.factor (y))   ## response variable should be factor
	install.packages("e1071")
	library (e1071)
	svmfit=svm(y~., data=dat, kernel ="linear", cost =10,scale =FALSE )   ## scale=FALSE: do not standardize the variable.

	plot(svmfit, dat)   ## support vectors are plotted as crosses.
	svmfit$index	  ## get the index of the support vectors
	summary (svmfit )   ## get basic information about support vector classifier

	svmfit =svm(y~., data=dat , kernel ="linear", cost =0.1,scale =FALSE )   ## change the cost
	plot(svmfit , dat)
	svmfit$index
	summary(svmfit)     ## there are support vectors 

	## use cross-validation to tune the parameter C

	set.seed (1)
	tune.out=tune(svm,y~.,data=dat ,kernel ="linear",ranges =list(cost=c(0.001 , 0.01, 0.1, 1,5,10,100)))  ## 10-fold CV
	summary (tune.out)	


	bestmod =tune.out$best.model   ## output the best model
	summary (bestmod )



	### generate new data and use SVM to predict response
	xtest=matrix (rnorm (20*2) , ncol =2)
	ytest=sample (c(-1,1) , 20, rep=TRUE)
	xtest[ytest ==1 ,]= xtest[ytest ==1,] + 1
	testdat =data.frame (x=xtest, y=as.factor(ytest))

	ypred=predict (bestmod ,testdat )
	table(predict =ypred , truth= testdat$y )


	## use cost=0.01 as a comparison
	svmfit =svm(y~., data=dat , kernel ="linear", cost =.01,scale =FALSE )
	ypred=predict (svmfit ,testdat )
	table(predict =ypred , truth= testdat$y )



	### consider the case the data are perfectly separable
	x[y==1 ,]= x[y==1 ,]+0.5
	plot(x, col =(y+5) /2, pch =19)


	dat=data.frame(x=x,y=as.factor (y))
	svmfit =svm(y~., data=dat , kernel ="linear", cost =1e5)
	summary (svmfit )
	plot(svmfit , dat)


	svmfit =svm(y~., data=dat , kernel ="linear", cost =1)  ## use a smaller cost value
	summary (svmfit )
	plot(svmfit ,dat )








	#####################
	#### Support Vector Machine
	##########################

	## fit SVM, we use kernel="radial" or "polynomial"

	## generate data with non-linear separation boundary

	set.seed (1)
	x=matrix (rnorm (200*2) , ncol =2)
	x[1:100 ,]=x[1:100 ,]+2
	x[101:150 ,]= x[101:150 ,] -2
	y=c(rep(1 ,150) ,rep (2 ,50) )
	dat=data.frame(x=x,y=as.factor (y))
	plot(x, col=y)


	train=sample (200 ,100)
	svmfit =svm(y~., data=dat [train ,], kernel ="radial", gamma =1,cost =1)  ## fit SVM with radial kernel
	plot(svmfit , dat[train ,])


	summary (svmfit )


	svmfit =svm(y~., data=dat [train ,], kernel ="radial",gamma =1,cost=1e5)  ## change the cost
	plot(svmfit ,dat [train ,])

	## tune(): to select the best cost and gamma using cross-validation
	set.seed (1)
	tune.out=tune(svm , y~., data=dat[train ,], kernel ="radial",ranges =list(cost=c(0.1 ,1 ,10 ,100 ,1000),gamma=c(0.5,1,2,3,4) ))
	summary (tune.out)

	## confusion matrix of the best model
	table(true=dat[-train ,"y"], pred=predict (tune.out$best.model ,newx=dat[-train ,]))


	## produce ROC curve

	install.packages("ROCR")
	library (ROCR)
	rocplot =function (pred , truth){
		predob = prediction(pred , truth)
		perf = performance (predob , "tpr", "fpr")
		plot(perf)}

	## ROC on training data
	svmfit.opt=svm(y~., data=dat[train ,], kernel ="radial",gamma =2, cost=1, decision.values =T)  ###decision.values =T: show predicted or fitted values: hat{f}
	fitted =attributes(predict(svmfit.opt ,dat[train ,], decision.values =TRUE))$decision.values  ## use predict to output fitted values
	par(mfrow =c(1,2))
	rocplot(fitted,dat[train ,"y"])  ## producde ROC curve


	## change gamma to 50
	svmfit.flex=svm (y~., data=dat[train ,], kernel ="radial",gamma =50, cost=1, decision.values =T)
	fitted =attributes (predict (svmfit.flex ,dat[train ,], decision.values =T))$decision.values
	rocplot (fitted,dat[train ,"y"])



	## ROC curve on test data set
	fitted =attributes (predict (svmfit.opt ,dat[-train ,], decision.values =T))$decision.values
	rocplot (fitted ,dat [-train ,"y"])
	fitted =attributes (predict (svmfit.flex ,dat[-train ,], decision.values =T))$decision.values
	rocplot (fitted ,dat [-train ,"y"])




	## SVM on multiple classes
	set.seed (1)
	x=rbind(x, matrix (rnorm (50*2) , ncol =2))
	y=c(y, rep (0 ,50) )
	x[y==0 ,2]= x[y==0 ,2]+2
	dat=data.frame(x=x, y=as.factor (y))
	par(mfrow =c(1,1))
	plot(x,col =(y+1))

	svmfit =svm(y~., data=dat , kernel ="radial", cost =10, gamma =1)
	plot(svmfit , dat)



	######################
	## SVM application to gene data
	#####################################


	## use Khan data set:  y has 4 valudes presenting 4 tumors. x are gene measurement
	library (ISLR)
	names(Khan)
	dim( Khan$xtrain )
	dim( Khan$xtest )
	length (Khan$ytrain )
	length (Khan$ytest )
 
	table(Khan$ytrain )
	table(Khan$ytest )


	dat=data.frame(x=Khan$xtrain , y=as.factor( Khan$ytrain ))
	out=svm(y~., data=dat , kernel ="linear",cost =10)
	summary (out)

	table(out$fitted , dat$y)  ## confusion matrix, diagonal elements represent number of observations correctly classified

	## confusion matrix on test data set
	dat.te=data.frame(x=Khan$xtest , y=as.factor (Khan$ytest ))
	pred.te=predict (out , newdata =dat.te)
	table(pred.te , dat.te$y)






























```
